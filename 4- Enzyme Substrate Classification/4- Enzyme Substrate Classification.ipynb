## x and y

```python
data
```

```python
x1 = data.drop(['ec1', 'ec2'], axis = 1)
x2 = data.drop(['ec1', 'ec2'], axis = 1)
y1 = data['ec1']
y2 = data['ec2']
```

```python
## test and train
from sklearn.model_selection import train_test_split
x1_train, x1_test, y1_train, y1_test = train_test_split(x1, y1, test_size = 0.2, random_state = 42)
x2_train, x2_test, y2_train, y2_test = train_test_split(x2, y2, test_size = 0.2, random_state = 42)
```

```python
## ec1 için svm, ec2 için bernoulli kullanıcaz.
from sklearn.svm import SVC
from sklearn.naive_bayes import BernoulliNB
model = SVC()
model.fit(x1_train, y1_train)
tahmin1 = model.predict(x1_test)
model2 = BernoulliNB()
model2.fit(x2_train, y2_train)
tahmin2 = model2.predict(x2_test)
```

```python
tahmin1
```

```python
tahmin2
```

```python
c = pd.DataFrame(tahmin2)
c.value_counts()
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(tahmin1, y1_test) * 100
```

```python
accuracy_score(tahmin2, y2_test) * 100
```
